### POS Tagging

```python
# English text
print(*[f'word: {word.text}\tupos: {word.upos}\txpos: {word.xpos}\tfeats: {word.feats if word.feats else "_"}' for sent in doc1.sentences for word in sent.words], sep='\n')
```

```python
pos_dict_en = {}

for sent in doc1.sentences:
    for word in sent.words:
        pos_dict_en.setdefault(str(word.upos),0)
        for key in pos_dict_en.keys():
            if key == str(word.upos):
                pos_dict_en[str(word.upos)] += 1

pos_dict_en
```

```python
# define function to find perc. error

def evaluate_percentage_error(list1, original_tag_lst):
    counts = 0
    for sent, tags in zip(list1, original_tag_lst):
        for word, tag in zip(sent.words, tags):
            if str(tag) == str(word.xpos)  or (str(tag) == 'SYM' and str(word.upos) == 'PUNCT') or ( str(tag) == 'DET' and str(word.xpos) == 'DT'):
                counts += 0
            else:
                print('Original:', str(tag),'|', 'Word:', str(word), '\n')
                counts += 1
    return print('Total percentage of error:',counts/len([item for sublist in original_tag_lst for item in sublist])*100, '%')
```

```python
# Define a function to find percentage of error between tagging of nltk and mine
# http://ucrel.lancs.ac.uk/bnc2/bnc2guide.htm -- VVD - VBD 
counts = 0
for to, ta in zip(token_list_spacy, [item for sublist in original_tags for item in sublist]):
    if to[1] == ta  or (ta == 'SYM' and to[1] in (',','!','.','``','',"''",'"','_SP') or (ta == 'DET' and to[1] =='DT')): 
        counts +=1
    else:
        print('FALSE:- ', to[0], to[1],ta)

print('Total percentage of error:', (1- counts/len([item for sublist in original_tags for item in sublist]))*100, '%')
```
