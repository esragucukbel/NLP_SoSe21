```python
import re
from nltk.util import ngrams

def create_n_grams(source_text, num):
    n_grams = list(ngrams(source_text, num))
    return ngrams, [ ' '.join(grams) for grams in n_grams]
```

```python
create_n_grams(bible,5)[1]
```

```python
def ngrams_model(source_text, num, word):
    for item in nltk.FreqDist(ngrams(source_text, num)):
        for token in item:
            if token == word:
                return print(item)
```

```python
ngrams_model(bible, 15, "pass")
```

## Exercise2 : wn meronyms

```python
from nltk.corpus import wordnet as wn
from nltk.corpus import brown
```

```python
tagged_religion_text = brown.tagged_sents(categories = 'religion')
```

```python
tagged_romance_text = brown.tagged_sents(categories = 'romance')
```

```python
def find_meronyms(word):
    count = 0
    for item in wn.synsets(word):
        count += len(item.part_meronyms())
        count += len(item.substance_meronyms())
    return count
```

```python
religion_meronyms_count = 0

for sentences in tagged_religion_text:
    for token in sentences:
        religion_meronyms_count += find_meronyms(token[0])
```

```python
religion_meronyms_count
```

```python
romance_meronyms_count = 0

for sentences in tagged_romance_text:
    for token in sentences:
        romance_meronyms_count += find_meronyms(token[0])

romance_meronyms_count
```

## Exercise3 : wn polysemy

Polysemy: The coexistence of many possible meanings for a word or phrase.
Ambiguity:  is a term used to characterise phenomena that have more than only one meaning. These meanings are distinct from each other and have no close schema in common.

```python
def polysemy(word):
    return len(wn.synsets(word))
```

#### Religion

```python
religion_polysemy = []

for sentences in tagged_religion_text:
    for token in sentences:
        religion_polysemy.append(polysemy(token[0]))
```

```python
print('Religion polysemy:', sum(religion_polysemy)/len(religion_polysemy))
```

#### Romance

```python
romance_polysemy = []

for sentences in tagged_romance_text:
    for token in sentences:
        romance_polysemy.append(polysemy(token[0]))
```

```python
print('Romance polysemy:', sum(romance_polysemy)/len(romance_polysemy))
```

```python
dog = wn.synset('dog.n.01')
cat = wn.synset('cat.n.01')
```

Return a score denoting how similar two word senses are, based on the shortest path that connects the senses in the is-a (hypernym/hypnoym) taxonomy. The score is in the range 0 to 1. By default, there is now a fake root node added to verbs so for cases where previously a path could not be found---and None was returned---it should return a value. The old behavior can be achieved by setting simulate_root to be False. A score of 1 represents identity i.e. comparing a sense with itself will return 1.

```python
wn.path_similarity(dog, cat)
```

Leacock-Chodorow Similarity: Return a score denoting how similar two word senses are, based on the shortest path that connects the senses (as above) and the maximum depth of the taxonomy in which the senses occur. The relationship is given as -log(p/2d) where p is the shortest path length and d the taxonomy depth.

```python
wn.lch_similarity(dog, cat)
```

Wu-Palmer Similarity: Return a score denoting how similar two word senses are, based on the depth of the two senses in the taxonomy and that of their Least Common Subsumer (most specific ancestor node). Note that at this time the scores given do _not_ always agree with those given by Pedersen's Perl implementation of Wordnet Similarity.

```python
wn.wup_similarity(dog, cat)
```

Resnik Similarity: Return a score denoting how similar two word senses are, based on the Information Content (IC) of the Least Common Subsumer (most specific ancestor node). Note that for any similarity measure that uses information content, the result is dependent on the corpus used to generate the information content and the specifics of how the information content was created.

```python
from nltk.corpus import wordnet_ic

brown_ic = wordnet_ic.ic('ic-brown.dat')
cat.res_similarity(dog, brown_ic)
```

Jiang-Conrath Similarity Return a score denoting how similar two word senses are, based on the Information Content (IC) of the Least Common Subsumer (most specific ancestor node) and that of the two input Synsets. The relationship is given by the equation 1 / (IC(s1) + IC(s2) - 2 * IC(lcs)).

```python
cat.jcn_similarity(dog, brown_ic)
```

Lin Similarity: Return a score denoting how similar two word senses are, based on the Information Content (IC) of the Least Common Subsumer (most specific ancestor node) and that of the two input Synsets. The relationship is given by the equation 2 * IC(lcs) / (IC(s1) + IC(s2)).

```python
cat.lin_similarity(dog, brown_ic)
```
